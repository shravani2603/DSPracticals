```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv("Iris.csv", header=None)
df
```

```python
col= ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm", "Species"]
df.columns=col
df
```

# 1.Create data subsets for different species

```python
df["Species"].unique()
```

```python
df_setosa= df[df["Species"]=="Iris-setosa"]
df_setosa.head()
```

```python
df_versicolor= df[df["Species"]=="Iris-versicolor"]
df_versicolor.head()
```

```python
df_virginica= df[df["Species"]=="Iris-virginica"]
df_virginica.head()
```

# 2.Merge two subsets

```python
df_setosa_versicolor= pd.concat([df_setosa, df_versicolor])
df_setosa_versicolor
```

```python
print("setosa shape: ", df_setosa.shape)
print("versicolor shape: ", df_versicolor.shape)
print("merge shape: ", df_setosa_versicolor.shape)
```

```python
df_merged= pd.concat([df_setosa, df_versicolor, df_virginica])
df_merged
```

```python
print("setosa shape: ", df_setosa.shape)
print("versicolor shape: ", df_versicolor.shape)
print("virginica shape: ", df_virginica.shape)
print("merge shape: ", df_merged.shape)
```

# 3. Sort Data Petal Length

```python
df_sort= df.sort_values(by=["PetalLengthCm"])
df_sort
```

# 4. Transposing Data

```python
df_transpose= df.transpose()
df_transpose
```

# 5.Melting Data to long format

```python
df_melted= pd.melt(df, id_vars=["SepalLengthCm", "PetalLengthCm", "Species"], var_name="Metric", value_name="value")
df_melted
```

```python
df_melted1= pd.melt(df, id_vars=["SepalLengthCm", "PetalLengthCm"], var_name="Metric", value_name="value")
df_melted1
```

# 6.Casting data to wide format

```python
df_casted= pd.pivot_table(df, index=["Species"])
df_casted
```

```python
df_casted1= pd.pivot_table(df, index=["Species", "PetalLengthCm"], values=["PetalWidthCm"])
df_casted1
```

```python
df_casted2= pd.pivot_table(df, index=["Species", "PetalLengthCm"])
df_casted2
```
