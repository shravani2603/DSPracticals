# 1.Create data subsets for different languages(Original Language). 

```python
df["original_language"].unique()
```

```python
df_en= df[df["original_language"]=="en"]
df_en.head()
```

```python
df_fr= df[df["original_language"]=="fr"]
df_fr.head()
```

```python
df_zh=df[df["original_language"]=="zh"]
df_zh.head()
```

```python
df_it= df[df["original_language"]=="it"]
df_it.head()
```

```python
df_fa= df[df["original_language"]=="fa"]
df_fa.head()
```

```python
df_nl= df[df["original_language"]=="nl"]
df_nl.head()
```

```python
df_da=df[df["original_language"]=="da"]
df_da.head()
```

```python
df_tl=df[df["original_language"]=="tl"]
df_tl.head()
```

# 2.Merge Two Subsets

```python
df_en_fr= pd.concat([df_en, df_fr])
df_en_fr
```

```python
print("Shape of en subset: ", df_en.shape)
print("Shape of fr subset: ", df_fr.shape)
print("Shape of merged subsets: ", df_en_fr.shape)
```

```python
df_zh_it= pd.concat([df_zh, df_it])
df_zh_it
```

```python
print("Shape of zh subset: ", df_zh.shape)
print("Shape of it subset: ", df_it.shape)
print("Shape of zh_it subset: ", df_zh_it.shape)
```

```python
df_fa_nl_da= pd.concat([df_fa, df_nl, df_da])
df_fa_nl_da
```

```python
print("Shape of fa subset: ", df_fa.shape)
print("Shape of nl subset: ", df_nl.shape)
print("Shape of da subset: ", df_da.shape)
print("Shape of fa_nl_da subset: ", df_fa_nl_da.shape)
```

# 5. Melting Data to Long Format

```python
df_melted= pd.melt(df, id_vars=['id', 'original_language'], var_name='Metric', value_name='value')
df_melted
```

```python
df_casted= pd.pivot_table(df, index=["id", "original_language"], values=["original_title", "status" ], aggfunc="count")
df_casted
```
