# 6.Casting data to wide format

```python
df_casted= pd.pivot_table(df, index=["workclass", "occupation"], values=["hours-per-week"])
df_casted


#This code creates a pivot table from the DataFrame df, where:

#The rows of the pivot table are indexed by unique combinations of the workclass an occupation columns from the original DataFrame.
#The values in the pivot table are obtained from the hours-per-week column of the original DataFrame.
#By default, the pivot table will calculate the mean (average) for 'hours-per-week' for each combination of 'workclass' and 'occupation'.
#So, the resulting pivot table provides a summary of the average number of hours per week worked for each combination of 'workclass' and 'occupation' in the original DataFrame.





```
