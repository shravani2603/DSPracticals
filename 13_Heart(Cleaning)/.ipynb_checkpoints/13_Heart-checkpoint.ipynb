```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv("Cleavland.csv", header=None)
df
```

```python
cols= ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach",	"exang", "oldpeak",	"slope", "ca", "thal", "num"]
df.columns = cols
df
```

# 1.Data cleaning(Remove NA, ?, Negative values etc.)

## Remove NA

```python
df.isnull().sum()
```

```python
df1= df.dropna()
df1
```

## Remove ?

```python
(df1.values=='?').sum()
```

```python
df2= df1.replace('?', np.NaN)
df2
```

```python
df2=df2.dropna()
```

```python
(df2.values=='?').sum()
```

## Remove Negative values

```python
df2= df2.astype(float)
```

```python
(df2<0).sum().sum()
```

```python
df= df2[df2>=0]
df
```

# 2.Error correcting(Outlier detection and removal)


```python
df.std()
```

```python
df.mean()
```

```python
df.shape
```

```python
#calculating z-score

# The Z-score, also known as the standard score, is a statistical measure that quantifies 
# how many standard deviations a data point is away from the mean of a distribution. 
# It is a way to standardize and compare values across different distributions.

def remove_outliers_zscore(data, threshold=3):
    z_scores= np.abs((data-data.mean())/ data.std())
    out=z_scores > 3
    print(out.sum())
    data = data[~out.any(axis=1)]
    return data
filtered_data= remove_outliers_zscore(df)
```

```python
filtered_data.shape
```

# 3.Data Transformation

```python
#First transforming sex 0-1 to F-M
df.loc[df["sex"]==1, "sex"]="M"
df.loc[df["sex"]==0, "sex"]="F"
df
```

```python
lb=LabelEncoder()
df["sex"]=lb.fit_transform(df["sex"])
df
```

# 4.Build Data model using regression and kNN methods and compare accuracy of heart disease prediction.

```python
X=df.drop("sex", axis=1)
Y=df["sex"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

## Logistic Regression

```python
reg=LogisticRegression()
reg.fit(X_train, Y_train)
```

```python
y_pred_reg= reg.predict(X_test)
accuracy_reg= accuracy_score(Y_test, y_pred_reg)
print("Accuracy of Logistic Regression: ",accuracy_reg)
```

## K-Nearest Neighbour

```python
knn=KNeighborsClassifier()
knn.fit(X_train, Y_train)
```

```python
y_pred_knn= knn.predict(X_test)
accuracy_knn= accuracy_score(Y_test, y_pred_knn)
print("Accuracy of KNN: ",accuracy_knn)
```
