```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Iris.csv", header=None)
df
```

```python
cols= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']
df.columns=cols
df
```

# 1.Data cleaning(Remove NA, ?, Negative values etc.)

## Remove NA

```python
df.isnull().sum()
```

```python
df=df.dropna()
df
```

## Remove ?

```python
(df.values=='?').sum()
```

```python
df.dropna()
```

# 3. Data transformation

```python
df["Species"].unique()
```

```python
lb=LabelEncoder()
df["Species"]= lb.fit_transform(df["Species"])
df
```

# 2. Error correcting(Outlier detection and removal)

```python
df.shape
```

```python
#Z-score
def remove_outliers_zscore(data, threshold=3):
    z_scores= np.abs((data-data.mean())/ data.std())
    out=z_scores > 3
    print(out.sum())
    data = data[~out.any(axis=1)]
    return data
filtered_data= remove_outliers_zscore(df)
```

```python
filtered_data.shape
```

# 4.Build Data model using regression and Naïve Bayes methods and compare accuracy of Iris Species Prediction.

```python
X= df.drop("Species", axis=1)
Y=df["Species"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
#logistic regression
reg=LogisticRegression()
reg.fit(X_train, Y_train)
```

```python
y_pred_reg= reg.predict(X_test)
accuracy_reg= accuracy_score(Y_test, y_pred_reg)
print("Accuracy of logistic regression: ", accuracy_reg)
```

```python
#Naive Bayes
nb= GaussianNB()
nb.fit(X_train, Y_train)
```

```python
y_pred_nb= nb.predict(X_test)
accuracy_nb= accuracy_score(Y_test, y_pred_nb)
print("Accuracy of Naive Bayes: ", accuracy_nb)
```
