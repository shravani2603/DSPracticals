```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv("BreastCancerWc.csv", header=None)
df
```

```python
cols =['Sample code number', 'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape','Marginal Adhesion','Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin','Normal Nucleoli','Mitoses','Class']
df.columns= cols
df
```

# 1.Data cleaning(Remove NA, ?, Negative values etc.)

```python
df.dropna()
```

```python
(df.values=='?').sum()
```

```python
df=df.replace('?', np.NaN)
df
```

```python
df.dropna(inplace=True)
```

# 2. Error correcting(Outlier detection and removal)

```python
df.shape
```

```python
df=df.astype(float)
```

```python
def remove_outliers_zscore(data, threshold=3):
    z_scores= np.abs((data - data.mean())/data.std())
    out= z_scores>3
    print(out.sum())
    data=data[~out.any(axis=1)]
    return data
filtered_data= remove_outliers_zscore(df)
```

# 3. Data transformation

```python
df["Class"].unique()
```

```python
df.loc[df["Class"]==2, "Class"]="Y"
df.loc[df["Class"]==4, "Class"]="N"

df.head()
```

```python
lb= LabelEncoder()
df["Class"]= lb.fit_transform(df["Class"])
df
```

# 4. Build Data model using regression and Naïve Bayes methods and compare accuracy of benign and malignant tumors in Breast Cancer Dataset.

```python
df.dropna()
```

```python
X= df.drop("Class", axis=1)
Y=df["Class"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
#Logistic Regression
reg= LogisticRegression()
reg.fit(X_train, Y_train)
```

```python
y_pred_reg= reg.predict(X_test)
accuracy_reg= accuracy_score(Y_test, y_pred_reg)
print("Accuracy LR: ", accuracy_reg)
```

```python
#Naive Bayes
nb= GaussianNB()
nb.fit(X_train, Y_train)
```

```python
y_pred_nb= nb.predict(X_test)
accuracy_nb= accuracy_score(Y_test, y_pred_reg)
print("Accuracy NB: ", accuracy_nb)
```
