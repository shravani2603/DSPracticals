# 2. Error correcting(Outlier detection and removal)

```python
df.shape
```

```python
def remove_outliers_zscore(data, threshold=3):
    numeric_data= data.select_dtypes(include=np.number)
    zscore= np.abs((numeric_data- numeric_data.mean())/ numeric_data.std())
    out= zscore>3
    print(out.sum())
    data= data[~out.any(axis=1)]
    return data
filtered_data= remove_outliers_zscore(df)
```

# 4.Build Data model using regression and NaÃ¯ve Bayes methods for prediction of income category (>=50k or <=50k) and compare accuracy Prediction.

```python
X= df.drop("target", axis=1)
Y=df["target"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# df_encoded = pd.get_dummies(df)
```

```python
#Logistic Regression
reg= LogisticRegression()
reg.fit(X_train, Y_train)
```

```python
y_pred_reg= reg.predict(X_test)
accuracy_reg= accuracy_score(Y_test, y_pred_reg)
print("ACCURACY reg: ", accuracy_reg)
```

```python
#Naibe Bayes
nb=GaussianNB()
nb.fit(X_train, Y_train)
```

```python
y_pred_nb= nb.predict(X_test)
acc_nb= accuracy_score(Y_test, y_pred_nb)
print("accuracy nb:", acc_nb)
```
