```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Cleavland.csv", header= None)
df
```

```python
df.columns = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num']
df.head()
```

```python
df.isna().sum().sum()
```

```python
(df.values=='?').sum().sum()
```

```python
df=df.replace('?', np.NaN)
```

```python
df.dropna(inplace=True)
```

```python
(df.values=='?').sum().sum()
```

# 1.Histograms

```python
plt.figure(figsize=(8,6))
plt.hist(df["age"], bins=10, color="skyblue")
plt.xlabel("age")
plt.ylabel("freq")
plt.title("age distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
df["sex"].value_counts().plot(kind="pie", autopct="%1.1f%%", colors=["lightblue", "lightpink"])
plt.ylabel("")
plt.title("Sex Distribution")
plt.show()
```

# 3.Boxplots

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="cp", y="chol", data=df)
plt.xlabel("cp")
plt.ylabel("chol")
plt.title("cp vs chol")
```

# 4.Scatterplots

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="age", y="thalach", data=df, hue="num")
plt.xlabel("age")
plt.ylabel("heart rate")
plt.title("age vs heart rate")
plt.show()
```

```python
fig,axes =plt.subplots(1,3,figsize=(10,5))
sns.boxplot(x='chol',data=df,ax=axes[0])
axes[0].set_title('Box Plot')
sns.histplot(data=df['chol'],ax=axes[1])
axes[1].set_title('Histogrm')
sns.scatterplot(x='chol',y='thal',data=df,ax=axes[2])
axes[2].set_title('scatter plot')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='chol', y='thalach', data=df, hue='cp')  # Assuming 'cp' represents different categories
plt.xlabel('Cholesterol')
plt.ylabel('Maximum Heart Rate')
plt.title('Scatterplot with Boxplots')

# Overlay boxplots for 'chol' on the same plot
sns.boxplot(x='chol', data=df, color='skyblue', width=0.2)

# Show the plot
plt.legend(title='Chest Pain Type')  # Add a legend for the scatterplot
plt.tight_layout()
plt.show()
```
