```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("adult.csv", header=None)
df
```

```python
df.columns=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','target']
df.head()
```

```python
data=df
```

# HISTOGRAM

```python
plt.figure(figsize=(8,6))
plt.hist(data['age'],color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

# DOTPLOTS

```python
plt.plot(df['fnlwgt'],'ro')
```

# 3.BARPLOTS

```python
plt.figure(figsize=(8,6))
df["age"].value_counts().plot(kind="bar", color="skyblue")
plt.xlabel("age")
plt.ylabel("freq")
plt.title("age vs freq")
plt.show()
```

```python
plt.figure(figsize=(8,6))
df["education-num"].value_counts().plot(kind="bar", color="skyblue")
plt.xlabel("education-num")
plt.ylabel("freq")
plt.title("education-num vs freq")
plt.show()
```

# LINE CHARTS

```python
plt.figure(figsize=(8,6))
plt.plot(df["hours-per-week"], color= "purple")
plt.xlabel("index")
plt.ylabel("hours-per-week")
plt.title("hours-per-week trend")
plt.show()
```

# add histogram and scatterplot to boxplot

```python
fig, axes= plt.subplots(1,3, figsize=(10,5))
sns.boxplot(x="fnlwgt", data=df, ax=axes[0])
axes[0].set_title("Box")

sns.histplot(data=df["fnlwgt"], ax=axes[1])
axes[1].set_title("histo")

sns.scatterplot(x="fnlwgt", y="hours-per-week", data=df, ax=axes[2])
axes[2].set_title("scatter")

plt.tight_layout()
plt.show()
```
