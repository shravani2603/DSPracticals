```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("adult.csv", header=None)
df
```

```python
df.columns=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','target']
df.head()
```

# histogram

```python
plt.figure(figsize=(8,6))
plt.hist(df["age"], bins=10, color="pink")
plt.xlabel("age")
plt.ylabel("freq")
plt.title("age vs freq")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df["education-num"], bins=10, color="pink")
plt.xlabel("education-num")
plt.ylabel("freq")
plt.title("education-num vs freq")
plt.show()
```

# PIE CHARTS

```python
plt.figure(figsize=(8,6))
df["education"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.ylabel(" ")
plt.title("education")
plt.show()
```

```python
plt.figure(figsize=(8,6))
df["relationship"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.ylabel(" ")
plt.title("relationship")
plt.show()
```

# BOX PLOTS


```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df,x='target',y='fnlwgt')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df,x='hours-per-week',y='education-num')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="age", y="hours-per-week", data=df, hue="race")
plt.xlabel("age")
plt.ylabel("hours")
plt.title("age vs. hours")
plt.show()
```

# Add boxplots to a scatterplot

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x= "fnlwgt", y="hours-per-week", data=df, hue="sex")
plt.xlabel("fnlwgt")
plt.ylabel("hours")
plt.title("Scatter with box")
# plt.show()

sns.boxplot(x="fnlwgt", data=df, color="red", width= 0.2)
plt.tight_layout()
plt.show()
```
