```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("hepatitis.csv", header=None)
df
```

```python
cols=["Class", "Age", "Sex", "Steroid","Antivirals", "Fatigue", "Malaise", "Anorexia", "Liver big", "Liver firm", "Spleen Palpable", "Spiders", "Ascites", "Varices", "Bilirubin", "Alk Phosphate", "SGOT", "Albumin", "Protime", "Histology"]
df.columns=cols
df
```

# 1.Create data subsets for different sex.

```python
df["Sex"].unique()
```

```python
df_sex1= df[df["Sex"]==1]
df_sex1
```

```python
df_sex2= df[df["Sex"]==2]
df_sex2
```

# 2.Merge two subsets

```python
df_merged= pd.concat([df_sex1, df_sex2])
df_merged
```

# 3.Sort Data using age, SGOT, PROTIME.

```python
df_age_sort= df.sort_values(by=["Age"])
df_age_sort
```

```python
df_sgot_sort= df.sort_values(by=["SGOT"], ascending=False)
df_sgot_sort
```

```python
df_protime_sort= df.sort_values(by=["Protime"], ascending = False)
df_protime_sort
```

# 4. Transposing Data

```python
df_transpose= df.transpose()
df_transpose
```

# 5.Melting Data to long format

```python
df_melted= pd.melt(df, id_vars=["Class", "Age"], var_name="Metrics", value_name="values")
df_melted
```

# 6.Casting data to wide format

```python
df_casted= pd.pivot_table(df, index=["Class", "Age", "Sex"], values=["Histology", "Antivirals"])
df_casted
```
